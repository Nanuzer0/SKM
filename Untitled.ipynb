```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import special
from scipy import linalg
from scipy import interpolate
from numpy import random
%matplotlib inline
```

Стром $11$ точек, взяв
$$
x_i = \frac{i-1}{10}, \quad y_i = {\rm erf} x_i\quad (i=1,\dots,11),
$$
где
$$
{\rm erf} x = \frac{2}{\sqrt{\pi}} \int_0^x e^{-t^2} dt
$$
– *функция ошибок*.

```python
x=np.fromfunction(lambda self, x: (x-1)/10, (1,12), dtype=int)[0][1:]
y=special.erf(x)
```

```python
plt.plot(x, y, 'o')
xx = np.linspace(0, 1, 500)
#plt.plot(xx, special.erf(xx), '-')
pass
```

Иинтерполирующий многочлен 10-й степени.
$$
f(x)=a_0 x^{10} + a_1 x^{9} + \dots + a_{9} x + a_{10}.
$$

Cистема линейных уравнений для нахождения
коэффициентов интерполирующего многочлена $10$-й степени.


$$
\left\{
\begin{array}{l}
a_0 x_0^{10} + a_1 x_0^{9} + \dots + a_{9} x_0 + a_{10} = y_0, \\
a_0 x_1^{10} + a_1 x_1^{9} + \dots + a_{9} x_1 + a_{10} = y_1, \\
\dots \\
a_0 x_{10}^{10} + a_1 x_{10}^{9} + \dots + a_{9} x_{10} + a_{10} = y_{10} \\
\end{array}
\right.
$$



$$
W = 
\left(
\begin{array}{ccccc}
    x_0^{10} &     x_0^{9} & \dots &    x_0  & 1 \\
    x_1^{10} &     x_1^{9} & \dots &    x_1  & 1 \\
\dots \\
    x_{10}^{10} &     x_{10}^{9} & \dots &    x_{10}  & 1 \\
\end{array}
\right).
$$
$$
a = 
\left(
\begin{array}{c}
    a_0     \\
    a_1     \\
    \vdots  \\
    a_{10}     \\
\end{array}
\right),
\qquad
y =
\left(
\begin{array}{c}
    y_0     \\
    y_1     \\
   \vdots   \\
    y_{10}     \\
\end{array}
\right),
$$
$Wa = y$.

Число обусловленности $W$

```python
np.linalg.cond(W)
```

Решаем систему

```python
W=np.vander(x)
a=linalg.solve(W,y)
```

Значения коэффициентов многочлена ($a$)

```python
a
```

Чтобы проверить насколько достоверно полученное решение, я рассчитал среднее значение среди относительных ошибок, полученных 
при сравнении случайных векторов (10000) и их вычесленного значения

```python
n=10000
err=np.zeros(n)
for iter in range(n):
    x_prec = random.rand(11)
    b = W@x_prec
    x_comp = linalg.solve(W, b)
    err[iter]=np.floor(np.log10(linalg.norm(x_comp - x_prec)/linalg.norm(x_prec)))
```

```python
np.average(err)
```

Получается примерно 9 верных цифр

Табулирование интерполянта на $x$

Абсолютная ошибка в промежуточных точках. Различие начинается где-то после 16 знака после запятой

```python
#plt.plot(np.polyval(a, 0.6) - np.log(0.6), y, 'o')
err=np.zeros(11)
for i in range(11):
    err[i]=np.polyval(a, x[i]) - special.erf(x[i])
plt.plot(x, err, 'o')
```

Относительная ошибка

```python
err2=np.zeros(11)
for i in range(11):
    if special.erf(x[i])==err[i]==0:
         err2[i]=0
         continue
    err2[i] = err[i] / special.erf(x[i])
plt.plot(x, err2, 'o')
```

График зависимости чисел обусловленности от степени интерполяционного многочлена

```python
oarr=np.zeros(50)
for i in range(2,50):
    x=np.fromfunction(lambda self, x: (x-1)/10, (1,i), dtype=int)[0][1:]
    W=np.vander(x)
    oarr[i]=np.linalg.cond(W)
xx = np.array(range(2,50))
plt.plot(xx, oarr[2:], '-')
plt.xlabel('Степень многочлена')
plt.ylabel('Число обусловленности')
```
